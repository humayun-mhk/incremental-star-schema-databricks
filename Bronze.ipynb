```sql
%python
if spark.catalog.tableExists("datamodelling.default.source_data"):
    last_load = spark.sql("select max(order_date) from datamodelling.default.source_data").collect()[0][0]
```

```sql
%python
last_load 
```

```sql
%python
spark.sql(f"""
    select * from datamodelling.default.source_data
    where order_date > '{last_load}'
""").createOrReplaceTempView("Bronze_source")
```

```sql
select * from Bronze_source
```

```sql
create or replace table datamodelling.bronze.Bronze_table
as
select * from Bronze_source 
```
