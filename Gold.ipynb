## DimCustomer

```sql
create or replace table datamodelling.gold.DimCustomer
as
select
    *,
    row_number() over(order by customer_id) as DimCutomerKey
from 
(select 
  distinct(customer_id),
  customer_name,
  customer_email
from 
  datamodelling.silver.silver_table)
```

```sql
select * from datamodelling.gold.dimcustomer
```

## DimProduct

```sql
CREATE OR REPLACE TABLE datamodelling.gold.DimProduct
AS
SELECT
    ROW_NUMBER() OVER (ORDER BY product_id) AS DimProductKey,
    product_id,
    product_name,
    product_category
FROM (
    SELECT 
        DISTINCT(product_id),
        product_name,
        product_category
    FROM datamodelling.silver.silver_table
);
```

```sql
select * from datamodelling.gold.dimproduct
```

## DimDate

```sql
CREATE OR REPLACE TABLE datamodelling.gold.DimDate
AS
SELECT
    ROW_NUMBER() OVER (ORDER BY order_date) AS DimDateKey,
    order_date,
    YEAR(order_date)  AS year,
    MONTH(order_date) AS month,
    DAY(order_date)   AS day
FROM (
    SELECT 
        DISTINCT(order_date)
    FROM datamodelling.silver.silver_table
);
```

```sql
select * from datamodelling.gold.DimDate
```

##Dim Payement

```sql
CREATE OR REPLACE TABLE datamodelling.gold.DimPayment
AS
SELECT
    ROW_NUMBER() OVER (ORDER BY payment_type) AS DimPaymentKey,
    payment_type
FROM (
    SELECT 
        DISTINCT(payment_type)
    FROM datamodelling.silver.silver_table
);
```

```sql
select * from datamodelling.gold.DimPayment
```

## Dim Sales

```sql
CREATE OR REPLACE TABLE datamodelling.gold.DimSales
AS
SELECT
    ROW_NUMBER() OVER (ORDER BY country) AS DimSalesKey,
    country
FROM (
    SELECT 
        DISTINCT(country)
    FROM datamodelling.silver.silver_table
);
```

```sql
select * from datamodelling.gold.DimSales
```

## Fact Table

```sql
SELECT
    ds.DimSalesKey,
    dc.DimCutomerKey,
    dp.DimProductKey,
    dpay.DimPaymentKey,

    F.quantity,
    F.unit_price
FROM datamodelling.silver.silver_table F

LEFT JOIN datamodelling.gold.DimCustomer dc
    ON F.customer_id = dc.customer_id

LEFT JOIN datamodelling.gold.DimProduct dp
    ON F.product_id = dp.product_id

LEFT JOIN datamodelling.gold.DimPayment dpay
    ON F.payment_type = dpay.payment_type

LEFT JOIN datamodelling.gold.DimSales ds
    ON F.country = ds.country;
```
